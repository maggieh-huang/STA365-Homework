## Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

## Answer to Q2

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pymc
import arviz as az
```

```python
n = 100
true_theta, true_tau = 0,1
data_generating_mechanism = \
stats.norm(loc=true_theta, scale=true_tau**(-0.5))

x = data_generating_mechanism.rvs(size=n)
plt.hist(x, density=True)
x_grid = np.linspace(-5*true_tau**(-0.5),5*true_tau**(-0.5),1000)
plt.plot(x_grid,data_generating_mechanism.pdf(x_grid));
```

### $\theta$ ~ Normal and $\tau$ ~ Gamma

```python
conjugate_normal_gamma = pymc.Model()
    
with conjugate_normal_gamma:

    # Priors for unknown model parameters
    
    theta0 = 0  # prior belief regarding true_theta
    theta_prior_n = 1  # strength of prior belief as units of data
    tau0 = theta_prior_n*true_tau  # prior is defined not in terms of tau but amount of info we have of theta
    theta = pymc.Normal("theta", mu=theta0, sigma=tau0**(-0.5))
   
    tau_SS_prior = 1  # prior belief regarding true_sum_of_squares
    tau_prior_n = 1  # strength of prior belief as units of data
    tau = pymc.Gamma("tau", alpha=tau_prior_n/2, 
                            beta=tau_SS_prior/2)
    
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)
    
    idata = pymc.sample(draws=10000, chains=4, tune=100)
```

```python
#diagnostics
az.summary(idata, round_to=5)
```

```python
az.plot_trace(idata, combined=True);
```

```python
az.plot_forest(idata, var_names=["theta", "tau"], 
               combined=True, hdi_prob=0.95, r_hat=True);
```

```python
az.plot_energy(idata);
```

### Bayesian Inference for $\theta$ ~ Normal and $\tau$ ~ Gamma

The 94% confidence interval for $\theta$ is (-0.14727, 0.20167). <br>
The 94% confidence interval for $\tau$ is (0.86597,	1.48535).

### $\theta$ ~ Gamma and $\tau$ ~ Exponential

```python
conjugate_gamma_expon = pymc.Model()
    
with conjugate_gamma_expon:

    # Priors for unknown model parameters
   
    theta_SS_prior = 1  # prior belief regarding true_sum_of_squares
    theta_prior_n = 1  # strength of prior belief as units of data
    theta = pymc.Gamma("theta", alpha=theta_prior_n/2, 
                            beta=theta_SS_prior/2)
    
    tau_prior_lambda = 1
    tau = pymc.Exponential("tau", lam=tau_prior_lambda)
    
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)
    
    idata2 = pymc.sample(draws=10000, chains=4, tune=80)
```

```python
# diagnostics
az.summary(idata2, round_to=5)
```

```python
az.plot_trace(idata2, combined=True);
```

```python
az.plot_forest(idata2, var_names=["theta", "tau"], 
               combined=True, hdi_prob=0.95, r_hat=True);
```

```python
az.plot_energy(idata2);
```

### Bayesian Inference for $\theta$ ~ Gamma and $\tau$ ~ Exponential

The 94% credible interval for $\tau$ is (0.86771, 1.48667). <br>
The 94% credible interval for $\theta$ is (0.00000, 0.15484).

### $\theta$ ~ Exponential and $\tau$ ~ Uniform

```python
conjugate_expon_unif = pymc.Model()
    
with conjugate_expon_unif:

    # Priors for unknown model parameters
   
    theta_prior_lambda = 1
    theta = pymc.Exponential("theta", lam=theta_prior_lambda)
    
    tau = pymc.Uniform("tau", lower=0, upper=1)
    
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)
    
    idata3 = pymc.sample(draws=10000, chains=4, tune=50)
```

```python
# diagnostics
az.summary(idata3, round_to=5)
```

```python
az.plot_trace(idata3, combined=True);
```

```python
az.plot_forest(idata3, var_names=["theta", "tau"], 
               combined=True, hdi_prob=0.95, r_hat=True);
```

```python
az.plot_energy(idata3);
```

### Bayesian Inference for $\theta$ ~ Exponential and $\tau$ ~ Uniform

The 94% credible interval for $\tau$ is (0.72195, 0.99998). <br>
The 94% credible interval for $\theta$ is (0.00002, 0.17777).

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

## Answer to Q3

In the code provided, the splice-sampling works using a two-step procedure and using another variable $y$. First, it finds an interval $[a, b]$ around the point $x$ where $f(x)>y$. $y$ is sampled uniformly from $y \sim  Unif(0, f(x^{(t)}_i|x_{-i}))$. The interval is extended such that both endpoints $f(a)$ and $f(b)$ are less than $y$. A candidate $x' \sim U(a,b) $ is drawn uniformly, and $f(x')$ is accepted as the next sample point if it is greater than $y$. Otherwise, the interval narrows depending on whether $x'$ is to the left or right of $x$, and the function is called again until another sample is obtained. 

In the Metropolis within Gibbs algorithm, full conditional distributions are usually only known up to a normalizing constant. Slice sampling can replace the Metropolis-Hastings step by proposing new values without needing to tune the proposal distributions. The curve shows us the unnormalized full conditional density. In slice sampling, the initial value for $x$ may be chosen from prior knowledge or another valid starting point. In Gibbs Sampling, we set initial values for all $x_i$'s, and for each $x_{i}$, we compute $f(x_{i}|x_{-i})$. We use slice sampling as shown above to update conditionally based on the values of the other variables. This process is iterated until convergence is obtained.
