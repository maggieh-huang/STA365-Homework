## Q1: derive the following

1. $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I)$ for the **linear regression model** likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$

2. The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

#### [Optional] Q1 Extra: derive the following

Show that for $s_i=1$ and **hyperparameters** $b_i=0$ (ignoring normalizing proportionality constants) the log posterior distributions for $\beta$ using either **normal** or **Laplace** prior distributions have analagous forms to the **ridge** and **lasso** loss functions.

*Note though that Bayesians do not optimize posterior distributions, they sample from them; but, nonetheless, the posterior distributions serve to provided 'regularizeations' of the likelihood through the prior.*

## Answer to Q1

### Part 1

$
\begin{align*}
p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma,\boldsymbol\beta_0, \Sigma_\beta) &\propto{}  p(\textbf{y}_{n\times 1}| \textbf{X}_{n\times p}, \boldsymbol \beta, \Sigma) \times p(\boldsymbol \beta_{p \times 1})\\
&={} (2\pi )^{-n/2}\det(\Sigma)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right) \times (2\pi )^{-p/2}\det(\Sigma_\beta)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\boldsymbol\beta - \boldsymbol\beta_0)^\top \Sigma_\beta^{-1}(\boldsymbol\beta - \boldsymbol\beta_0)\right)\\
&\propto{} \exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right) \times  \exp \left(-{\frac {1}{2}}(\boldsymbol\beta - \boldsymbol\beta_0)^\top \Sigma_\beta^{-1}(\boldsymbol\beta - \boldsymbol\beta_0)\right)\\
&\propto{} \exp \left(-{\frac {1}{2}}({\boldsymbol \beta^\top}\mathbf{X^\top}\Sigma^{-1} \mathbf{X\boldsymbol \beta} - 2{\boldsymbol \beta^\top}\mathbf{X^\top}\Sigma^{-1} \mathbf{y}) \right) \times \exp \left(-{\frac {1}{2}}(\boldsymbol\beta^\top \Sigma_\beta^{-1}\boldsymbol\beta - 2 \boldsymbol \beta^\top \Sigma_\beta^{-1} \boldsymbol\beta_0)\right) \\
&={} \exp \left(-{\frac {1}{2}}({\boldsymbol \beta^\top}\left(\mathbf{X^\top}\Sigma^{-1} \mathbf{X} + \Sigma_\beta^{-1} \right)\boldsymbol \beta - 2{\boldsymbol \beta^\top}\left(\mathbf{X^\top}\Sigma^{-1} \mathbf{y} + \Sigma_\beta^{-1} \boldsymbol\beta_0\right))\right) \\
\text{Substitute } A &= \mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma_\beta^{-1} \text{. A is the inverse of the covariance matrix and b is the linear term}\\
\text{Substitute } b &= \mathbf{X}^\top \Sigma^{-1} \mathbf{y} + \Sigma_\beta^{-1} \boldsymbol{\beta}_0 \\
&={} \exp \left(-{\frac {1}{2}}{\boldsymbol \beta^\top} A {\boldsymbol \beta} - 2{\boldsymbol \beta^\top}b\right) \\
\text{This is in quadratic form. Let } \mu &= A^{-1}b \text{. Completing the square, we get the following:}\\
&={} \exp \left( \frac{-1}{2} [({\boldsymbol \beta} - \mu)^{T} A ({\boldsymbol \beta} - \mu) - \mu^{T} A \mu ]\right) \\
&={} \exp \left( \frac{-1}{2} ({\boldsymbol \beta} - \mu)^{T} A ({\boldsymbol \beta} - \mu) \right) \\
\text{After undoing substitutions, the final answer is the following: } \\
\Longrightarrow p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma,\boldsymbol\beta_0, \Sigma_\beta) &={} 
\mathcal{MVN}\left(E[\boldsymbol \beta] = COV[\boldsymbol \beta] \left(\mathbf{X^\top}\Sigma^{-1} \mathbf{y} + \Sigma_\beta^{-1} \boldsymbol\beta_0\right),  COV[\boldsymbol \beta] = \left(\mathbf{X^\top}\Sigma^{-1} \mathbf{X} + \Sigma_\beta^{-1} \right)^{-1} \right)
\\
\end{align*}
$


### Part 2

In lectures notes, we have that $\Sigma=\sigma^{2} I$. Therefore, the likelihood becomes the following:

$
\begin{align*}
p(\textbf{y}| \textbf{X}, \boldsymbol \beta, \Sigma = \sigma^{2} I) &={} (2\pi )^{-n/2}\det(\Sigma)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)\\
p(\textbf{y}| \textbf{X}, \boldsymbol \beta, \sigma^{2}) &={} (2\pi )^{-n/2}\det(\sigma^2 I)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\sigma^2 I)^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)\\
p(\textbf{y}| \textbf{X}, \boldsymbol \beta, \sigma^{2}) &={} (2\pi )^{-n/2}\det(\sigma^2 I)^{-1/2}\,\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\sigma^2 I)^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)\\
p(\textbf{y}| \textbf{X}, \boldsymbol \beta, \sigma^{2}) &={} (2\pi )^{-n/2}(\sigma^2)^{-n/2}\,\exp \left(-{\frac {1}{2 \sigma^2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)\\
\end{align*}
$

The prior for $\sigma^2$ is given below: <br>
$
\begin{align*}
p(\sigma^{2}| \alpha^*, \beta^*) &={} \frac{\beta^{* \alpha^*}}{\Gamma(\alpha^*)} (\sigma^2)^{- \alpha^* - 1} e^{\frac{- \beta^*}{\sigma^2}}\\ 
\end{align*}
$

Therefore, the posterior analytical distribution for $\sigma^2$ is the following:

$
\begin{align*}
p(\sigma^{2}| y, \boldsymbol \beta, \textbf{X}, \alpha^*, \beta^*) &\propto{} \frac{\beta^{* \alpha^*}}{\Gamma(\alpha^*)} (\sigma^2)^{- \alpha^* - 1} e^{\frac{- \beta^*}{\sigma^2}} (2\pi )^{-n/2}(\sigma^2)^{\frac{n}{2}} e^{\left(-{\frac {1}{2 \sigma^2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)}\\ 
&\propto{} (\sigma^2)^{- \alpha^* - 1} e^{\frac{- \beta^*}{\sigma^2}} (\sigma^2)^{\frac{n}{2}} e^{\left(-{\frac {1}{2 \sigma^2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)}\\
&\propto{} (\sigma^2)^{- (\alpha^* + 1 + \frac{n}{2})} e^{\frac{- \beta^*}{\sigma^2} + \left(-{\frac {1}{2 \sigma^2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)}\\
&\propto{} (\sigma^2)^{- (\alpha^* + 1 + \frac{n}{2})} e^{\frac{-1}{\sigma^2} \left(\beta^* + {\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top (\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)}\\
\end{align*}
$

The posterior is
$p(\sigma^{2}| \mathbf{y}, \boldsymbol \beta, \mathbf{X}, \alpha^*, \beta^*) \sim \text{Inverse-Gamma} \left(\alpha^* + \frac{n}{2}, \beta^* + \frac{1}{2}(\mathbf{y} -\mathbf{X}\boldsymbol\beta)^\top (\mathbf{y} - \mathbf{X}\boldsymbol\beta)\right)$

```python
import pymc as pm
import numpy as np
import pandas as pd
import arviz as az
import seaborn as sns
from scipy import stats 
import matplotlib.pyplot as plt
```

### Dataset

Source: https://www.kaggle.com/datasets/hussainnasirkhan/multiple-linear-regression-dataset <br>
Variables: <br>
- Age (yrs): Independent variable
- Years of Experience: Independent variable
- Income: Dependent variable

```python
data = pd.read_csv('income.csv')
data
```

### Perform Bayesian Linear Regression

#### Preliminary Visualization of Data

First, we visualize the data. The independent variables are age and experience, while the dependent variable is income. After plotting age and experience against income, we see that the relationships are fairly linear and do not require any transformations. 

```python
plt.style.use('ggplot')

# subplots for histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].hist(data['age'], bins=15, color='skyblue', edgecolor='black')
axs[0].set_title('Distribution of Age')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')
axs[0].grid(True, linestyle='--', linewidth=0.5)

axs[1].hist(data['experience'], bins=15, color='lightgreen', edgecolor='black')
axs[1].set_title('Distribution of Experience')
axs[1].set_xlabel('Years of Experience')
axs[1].set_ylabel('Frequency')
axs[1].grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 5))
plt.scatter(data['age'], data['income'], alpha=0.7, color='coral', edgecolor='black')

# fit and plot regression line
z = np.polyfit(data['age'], data['income'], 1)
p = np.poly1d(z)
plt.plot(data['age'], p(data['age']), "r--")

plt.title('Income vs. Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()

# scatter plot for Experience vs. Income with regression line
plt.figure(figsize=(6, 5))
plt.scatter(data['experience'], data['income'], alpha=0.7, color='mediumpurple', edgecolor='black')

# fit and plot regression line
z = np.polyfit(data['experience'], data['income'], 1)
p = np.poly1d(z)
plt.plot(data['experience'], p(data['experience']), "r--")

plt.title('Income vs. Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Income')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

#### Use PyMC

```python
X = data[['age', 'experience']]
X = (X - X.mean()) / X.std()

y = data['income'].values
p = X.shape[1]

with pm.Model() as model:
    
    # priors for intercept
    alpha = pm.Normal('alpha', mu=0, sigma=10000)
    
    # priors for coefficients
    betas = pm.Normal('betas', mu=0, sigma=10, shape=p)
    
    # sigma
    sigma = pm.HalfNormal('sigma', sigma=10000)
    
    # linear model
    mu = alpha + pm.math.dot(X, betas)
    
    # likelihood
    y_obs = pm.Normal('y', mu=mu, sigma=sigma, observed=y)
    
    # sampling from posterior
    trace = pm.sample(2000, tune=1000, return_inferencedata=True)
```

#### Diagnostics and Posterior Predictive Checks

```python
az.plot_trace(trace)
plt.tight_layout()
plt.show()
```

```python
az.plot_forest(trace, var_names=["alpha", "betas", "sigma"], 
               combined=True, hdi_prob=0.95, r_hat=True);
```

```python
az.plot_energy(trace);
```

```python
with model:
    ppc = pm.sample_posterior_predictive(trace, random_seed=42)

az.plot_ppc(ppc)
plt.show()
```

#### Check Linear Regression Assumptions

```python
y_pred = ppc.posterior_predictive['y'].mean(dim=["chain", "draw"]).values
residuals = y - y_pred

# Residuals vs Fitted
sns.residplot(x=y_pred, y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.show()
```

```python
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('QQ Plot of Residuals')
plt.show()
```

#### Obtaining Bayesian Inference

The 94% credible intervals are displayed in the chart below:

```python
az.summary(trace, round_to=5)
```

#### Interpretation and Discussion

Overall, the model fits the data well. Trace plots demonstrate convergence and consistent mixing across the entire sample. The forest plot shows R-hat values at 1, further showing that the chains have converged. The HDI bars are also relatively narrow, demonstrating that the data strongly informs the parameters. The energy plot reveals good overlap between the theoretical and empirical energies. Through posterior predictive sampling, we observe that the model's predictions are close to the observed data's distribution and reproduce the overall shape and spread of the data. Finally, plots of residuals and the QQ plot show us that linear regression assumptions have been met.

#### Adding Outliers

Data points 20-23 were outliers that were manually added into the dataset.

```python
data2 = pd.read_csv('income_outlier.csv')
data2
```

#### Visualize Dataset

```python
plt.style.use('ggplot')

# subplots for histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Age histogram
axs[0].hist(data2['age'], bins=15, color='skyblue', edgecolor='black')
axs[0].set_title('Distribution of Age')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')
axs[0].grid(True, linestyle='--', linewidth=0.5)

# Experience histogram
axs[1].hist(data2['experience'], bins=15, color='lightgreen', edgecolor='black')
axs[1].set_title('Distribution of Experience')
axs[1].set_xlabel('Years of Experience')
axs[1].set_ylabel('Frequency')
axs[1].grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()

# scatter plot for Age vs. Income with regression line
plt.figure(figsize=(6, 5))
plt.scatter(data2['age'], data2['income'], alpha=0.7, color='coral', edgecolor='black')

# fit and plot regression line
z = np.polyfit(data2['age'], data2['income'], 1)
p = np.poly1d(z)
plt.plot(data2['age'], p(data2['age']), "r--")

plt.title('Income vs. Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()

# scatter plot for Experience vs. Income with regression line
plt.figure(figsize=(6, 5))
plt.scatter(data2['experience'], data2['income'], alpha=0.7, color='mediumpurple', edgecolor='black')

# fit and plot regression line
z = np.polyfit(data2['experience'], data2['income'], 1)
p = np.poly1d(z)
plt.plot(data2['experience'], p(data2['experience']), "r--")

plt.title('Income vs. Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Income')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

#### Modeling with PyMC

```python
X2 = data2[['age', 'experience']]
X2 = (X2 - X2.mean()) / X2.std()

y2 = data2['income'].values
p2 = X2.shape[1]

n = len(y2)
nu = 1

with pm.Model() as model:
    
    tau = pm.Gamma("tau", alpha = nu/2, beta = nu/2)
    
    tau_i = pm.Gamma("tau_i", alpha=nu/2, beta=nu/2, shape = n)
    
    alpha = pm.Normal('alpha', mu=0, sigma=10000)
    
    betas = pm.Normal('betas', mu=0, sigma=10, shape=p2)

    mu = alpha + pm.math.dot(X2, betas)
    
    # likelihood
    sigma_i = 1 / pm.math.sqrt(tau * tau_i)
        
    y2_obs = pm.Normal('y2', mu=mu, sigma=sigma_i, observed=y2)
    
    # sampling from posterior
    trace2 = pm.sample(2000, tune=1000, return_inferencedata=True)
```

#### Examining Outliers

```python
# posterior samples for tau_i from trace2
tau_i_samples = trace2.posterior['tau_i'].values  

# combine chains and draws
tau_i_samples_flat = tau_i_samples.reshape(-1, tau_i_samples.shape[-1]) 

tau_i_summary = az.summary(trace2, var_names=['tau_i'])
print(tau_i_summary)

plt.boxplot(tau_i_samples_flat)
plt.xlabel("Data Point Index")
plt.ylabel("Posterior of Tau_i")
plt.title("Identifying Outliers via Tau_i")
plt.show()
```

```python
X2_array = X2.values

tau_samples = trace2.posterior['tau'].values.reshape(-1)  
tau_i_samples = trace2.posterior['tau_i'].values.reshape(-1, n) 

# find sigma_i^{-2} for each posterior sample
sigma_inv_sq_samples = tau_samples[:, None] * tau_i_samples  

# find influence for each posterior sample
influences = []
for sample_idx in range(sigma_inv_sq_samples.shape[0]):
  
    D = np.diag(1 / sigma_inv_sq_samples[sample_idx])  
    
   
    XtDX_inv = np.linalg.inv(X2_array.T @ D @ X2_array)
    H = X2_array @ XtDX_inv @ X2_array.T @ D
    
   
    h_diag = np.diag(H)
    influences.append(h_diag)

influences = np.array(influences)  

influence_summary = {
    'mean': influences.mean(axis=0),
    'hdi_2.5%': np.percentile(influences, 2.5, axis=0),
    'hdi_97.5%': np.percentile(influences, 97.5, axis=0)
}

plt.boxplot([influences[:, i] for i in range(n)])
plt.xlabel("Data Point Index")
plt.ylabel("Influence (Hat Matrix Diagonal)")
plt.title("Posterior Influence of Each Data Point")
plt.show()
```

```python
outlier_indices = [20, 23] 
non_outlier_indices = [10, 20]  

# compare influence distributions
for idx in outlier_indices + non_outlier_indices:
    plt.hist(influences[:, idx], bins=30, alpha=0.5, label=f"Data Point {idx}")

plt.legend()
plt.xlabel("Influence (Hat Matrix Diagonal)")
plt.ylabel("Frequency")
plt.title("Influence Comparison: Outliers vs Non-Outliers")
plt.show()
```

```python
influence_df = pd.DataFrame({
    'Mean Influence': influence_summary['mean'],
    '2.5% HDI': influence_summary['hdi_2.5%'],
    '97.5% HDI': influence_summary['hdi_97.5%']
})
print(influence_df)
```
